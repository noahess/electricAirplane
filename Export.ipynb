```python
import json
import glob
```

```python
direct = glob.glob('Data/*.json')
```

```python
for file in direct:
    with open(file) as in_file:
        f = json.load(in_file)
    filename = f"Dats/{f['name']}-{'-'.join([str(x) for x in f['date'][0:3]])}.dat"
    with open(filename, 'w') as in_file:
        for datapt in f['data']:
            in_file.write('\t'.join([str(x) for x in datapt]))
            in_file.write('\n')
```
